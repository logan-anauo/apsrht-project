## **Antarctic Plateau Salt Relative Humidity and Temperature Dataset (APSRHT) - Cleaning**

Dataset: Davila, A; Parro, V. Antarctic Plateau-Salt Relative Humidity and Temperature. Astrobiology Habitable Environment Database. Retrieved Date: Jun 23, 2026. https://doi.org/10.48667/sy0j-yc84

"First column is the measurement number. Second column is the date/time of each measurement. Third column is the temperature in degrees Celssius. Fourth column is the RH as measured by the sensor. Fifth column is the RH with respect to ice"

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Antarctic_Plateau_Salt data.csv', encoding='latin1')
data.columns = data.iloc[0]
data = data[1:].reset_index(drop=True)
data
```

```python
### rename columns
data = data.rename(columns={'#': 'measurement_number', 'Date Time': 'date_time', 'T (Â¼C)': 'temp_C', 'RH (%)': 'relative_humidity', 'RH (ice) (%)': 'relative_humidity_corrected'})
data.head()
```

```python
### convert the column to datetime
data['date_time'] = pd.to_datetime(data['date_time'], format='%m/%d/%y %H:%M')
data.head()
```

```python
### set datetime as the dataframe index
data.set_index('date_time', inplace=True)
data.head()
```

```python
## create a complete 15-minute grid from start to end date
full_time_index = pd.date_range(start=data.index.min(), end=data.index.max(), freq='15min')
full_time_index
```

```python
### re-index dataframe to find gaps
data = data.reindex(full_time_index)
data.head()
```

```python
### date_time/index broken down into useful data columns
### 0=Monday, 1=Tuesday, 2=Wednesday, 3=Thursday, 4=Friday, 5=Saturday, 6=Sunday
data['year'] = data.index.year
data['month'] = data.index.month
data['day'] = data.index.day
data['hour'] = data.index.hour
data['day_of_week'] = data.index.dayofweek
data.head()
```

```python
### write to .csv
data.to_csv('apsrht_cleaned.csv', index=False, encoding='utf-8')
```
